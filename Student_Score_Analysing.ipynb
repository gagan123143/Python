## Analysing Scores of Students

```python
import numpy as np
from numpy import random
```

```python
stud_no=int(input("Enter The No Of Student:"))
```

```python
scores=np.random.randint(1,101,size=stud_no)
print(f"Scores:\n{scores}")
```

```python
grade_A=(scores>=90) 
grade_B=(scores>=80) & (scores<90)
grade_C=(scores>=70) &(scores<80)
grade_D=(scores>=60) &(scores<70)
grade_F=(scores<60)
```

```python
A=np.sum(grade_A)
B=np.sum(grade_B)
C=np.sum(grade_C)
D=np.sum(grade_D)
F=np.sum(grade_F)
print(f"No of Students with Grades:\nGrade-A: {A}\nGrade-B: {B}\nGrade-C: {C}\nGrade-D: {D}\nGrade-F: {F}")
```
